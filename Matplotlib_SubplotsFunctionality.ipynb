```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(4, 10), dpi=200)

axes[0][0].plot(x, y)

axes[0][1].plot(a, b)

axes[1][0].plot(x, y)
axes[1][0].set_ylabel("Y label 1,0")
axes[1][0].set_xlim(2, 6)

axes[1][1].plot(a, b)
axes[1][1].set_title('Title 1,1')

#fig.subplots_adjust(wspace=0.9, hspace=0.9)
fig.suptitle("Figure Title", fontsize=16)
plt.tight_layout()

fig.set_figwidth(10)

fig.savefig('new_subplots.png', bbox_inches='tight')
```
