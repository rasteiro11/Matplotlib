```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#linestyle options '--', '-', '-.', ':', 'steps'
ax.plot(x, x, color='#e6c017', label='X vs X', lw=4, ls='--') # RGB HEX Code
ax.plot(x, x+1, color='purple',label='X vs X+1', linewidth=2, linestyle='-.')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x, x+1, color='purple', lw=5)
lines[0].set_dashes([5, 2, 10, 2])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, color='purple', lw=2, marker='o', ls='--')
ax.plot(x, x+1, color='red', lw=2, marker='+', ls='--', ms=20, )
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, color='purple', lw=2, marker='o', ls='--', ms=10, markerfacecolor='red', markeredgewidth='4', markeredgecolor='orange')
```
